```python
train_set['Country'].value_counts().to_dict()
#Once i able to get this into dictionary I can map each with values we got like for United states we have frequency count=29170. This is wht Count of Frequency encoding
#Count of frquency encoding this is one of the encoding techniques.
#Now we need to understand how to replace this frequency encoding in 'Country' columsn
```

```python
type(train_set['Country'].value_counts().to_dict())
```

```python
country_map=train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

##### Advantages
1. Easy To Use
2. Not increasing feature space
##### Disadvantages
1. It will provide same weight if the frequencies are same.(If two countries of frequency are same so the will not be able to distinguish btw two countries). It will not be able to handle that kind of situation.

#### Target Guided Ordinal Encoding
1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

#### Mean Encoding

```python
df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
#Mean encoding provied monotonic relationship. It means as rank increase mean() values increases.
#One problem is tht it leads to overfitting.
```
